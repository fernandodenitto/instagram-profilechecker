# Logistic Regression with NaN Handling

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model.logistic import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
df_real = pd.read_csv('../data/data_real.csv')
df_fake = pd.read_csv('../data/data_fake.csv')
```

```python
df_real.head()
```

```python
df_real.describe()
```

```python
df_real.fillna(df_real.median(),inplace=True)
```

```python
df_fake.fillna(df_fake.median(),inplace=True)
df_fake.head()
```

```python
df=pd.concat([df_real,df_fake],ignore_index=True)
```

```python
print(df)
```

```python
# df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
features_columns=list(df.columns[:30])
# targets_column=list(df.columns[30:31])
# df.head()
print(features_columns)
```

```python
# cl_f = {'pos': 1, 'neg': 0}
# df['Class'] = df['Class'].map(cl_f)
```

```python
df.dropna(inplace=True)
df.shape
```

```python
targets = df['real_account'].values
```

```python
targets
```

```python
# df.drop('real_account', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
print(df)
```

```python
df_sc = StandardScaler().fit_transform(df)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_sc, targets, test_size = 0.99, random_state = 12345)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr_pred = lr.predict(x_test)
```

```python
lr.predict(x_test[14].reshape(1,-1))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, lr_pred))
```

```python
print(classification_report(y_test, lr_pred))
```
